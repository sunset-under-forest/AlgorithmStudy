```python
# a performance comparison between the min_max with and min_max_alpha_beta

from time import time
from minimax import *
```

```python
# draw the time spent and the recursion calls
import matplotlib.pyplot as plt
import numpy as np

time_spent_min_max = []
time_spent_min_max_alpha_beta = []
recursion_calls_min_max = []
recursion_calls_min_max_alpha_beta = []

for i in range(10):
    print(i)
    time_spent, recursion_calls = game(min_max)
    time_spent_min_max.append(time_spent)
    recursion_calls_min_max.append(recursion_calls)
    print(i)
    time_spent, recursion_calls = game(min_max_alpha_beta)
    time_spent_min_max_alpha_beta.append(time_spent)
    recursion_calls_min_max_alpha_beta.append(recursion_calls)

# draw the time spent
plt.figure(figsize=(10, 5))
plt.plot(time_spent_min_max, label='min_max')
plt.plot(time_spent_min_max_alpha_beta, label='min_max_alpha_beta')
plt.legend()
plt.show()
```

```python
# draw the recursion calls
plt.figure(figsize=(10, 5))
plt.plot(recursion_calls_min_max, label='min_max')
plt.plot(recursion_calls_min_max_alpha_beta, label='min_max_alpha_beta')
plt.legend()
plt.show()
```
